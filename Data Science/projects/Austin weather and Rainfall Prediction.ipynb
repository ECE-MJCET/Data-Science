```python
# Importing the pandas library for data manipulation and analysis
import pandas as pd

# Importing the pyplot module from the matplotlib library for creating plots and visualizations
import matplotlib.pyplot as plt

# Importing the LinearRegression class from scikit-learn
# Used for modeling relationships between dependent and independent variables
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("rainfall prediction.csv")          
df
```

```python
x = df.drop(['PrecipitationSumInches'], axis=1)
```

```python
# reshaping it into 2d vector
y = df['PrecipitationSumInches']
y = y.values.reshape(-1,1)
```

```python
day_index = 798
days = [i for i in range(y.size)]
```

```python
# Initializing a Linear Regression classifier
clf = LinearRegression()

# Creating an instance of the LinearRegression class from scikit-learn (sklearn)
# Training the classifier
clf.fit(x, y)
```

```python
print("The precipitation trend graph:")

# Create a scatter plot to visualize the precipitation trend over days
plt.scatter(days, y, color='g')  # Scatter plot of days vs. precipitation
plt.scatter(days[day_index], y[day_index], color='r')  # Highlight a specific day

# Set title, labels, and display the plot
plt.title("Precipitation level")
plt.xlabel("Days")
plt.ylabel("Precipitation in inches")

plt.show()

# Assuming x is your DataFrame containing attributes
# Filter the DataFrame x to select specific attributes for visualization
x_vis = x.filter(['TempAvgF', 'DewPointAvgF', 'HumidityAvgPercent', 'SeaLevelPressureAvgInches', 'VisibilityAvgMiles', 'WindAvgMPH'])
```

```python
import matplotlib.pyplot as plt

print("The precipitation vs attributes trend graph:")

# Assuming x_vis is your DataFrame containing attributes
# Assuming days and day_index are defined

for i in range(x_vis.columns.size):  # Loop through each column in the DataFrame
    plt.subplot(3, 2, i+1)  # Create subplots in a 3x2 grid
    plt.scatter(days, x_vis[x_vis.columns.values[i][:100]], color='g')  # Plot attribute trend over days
    plt.scatter(days[day_index], x_vis[x_vis.columns.values[i]][day_index], color='r')  # Highlight a specific day
    plt.title(x_vis.columns.values[i])  # Set subplot title to attribute name

plt.tight_layout()  # Adjust subplot layout to prevent overlapping
plt.show()  # Display the plot
```
